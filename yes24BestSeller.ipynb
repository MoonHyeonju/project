```python
from selenium import webdriver

# 1. 브라우저 열기
browser = webdriver.Chrome()
```

```python
# 2. 웹페이지 접속하기
url = 'https://www.yes24.com/product/category/daybestseller?categoryNumber=001&pageNumber=1&pageSize=24&type=day'
browser.get(url)
```

```python
# 3. 원하는 정보 있는지 확인하기
# 4. 브라우저가 가지고 있는 데이터(html) 다운&해석
from bs4 import BeautifulSoup
soup = BeautifulSoup(browser.page_source, "html.parser")
```

```python
# 5. 수집하려는 정보 정리하기
book_part_list = soup.select('#yesBestList > li') # ID 속성값이 yesBestList라는 태그 바로 아래에 있는 li 태그
len(book_part_list)
```

```python
# 책 1권 정보 수집/정리하기
book_part = book_part_list[0]
title = book_part.select('a.gd_name')[0].text
title
```

```python
results = []
for book_part in book_part_list:
    title = book_part.select('a.gd_name')[0].text
    company = book_part.select('span.authPub.info_pub > a')[0].text
    publish_date = book_part.select('span.authPub.info_date')[0].text
    auth = book_part.select('span.authPub.info_auth')[0].text.strip()
    print(title, company, publish_date, auth, sep=' / ')
    data = [title, company, publish_date, auth]
    results.append(data)
```

```python
import pandas as pd
df = pd.DataFrame(results)
df.columns = ['책제목', '출판사', '출판년월', '저자정보']
df.to_excel('./베스트셀러_index제외.xlsx', index=False)
```
