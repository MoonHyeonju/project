```python
import chromedriver_autoinstaller
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
chromedriver_autoinstaller.install()
```

```python
browser = webdriver.Chrome()
```

```python
url = 'https://www.melon.com/chart/index.htm'
browser.get(url)
```

```python
html = browser.page_source
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
lst = soup.select('tr.lst50')
lst
```

```python
len(lst)
```

```python
lst[-1]
```

```python
# 1위~50위: tr.lst50
# 51위~100위: tr.lst100

lst50 = soup.select('tr.lst50')
lst100 = soup.select('tr.lst100')

lst = lst50 + lst100
```

```python
len(lst)
```

```python
# 8. 정보 뭉치 1개에서 우리가 원하는 정보 찾기
element = lst[0]
element
```

```python
ranking = element.select('span.rank')
ranking # 리스트 타입의 객체
```

```python
# 리스트 타입의 객체에서 첫번째 원소 선택하기
ranking = element.select('span.rank')[0]
ranking
```

```python
ranking = element.select('span.rank')[0].text
ranking
```

```python
# 노래 제목 찾기 - a 태그에서 좀 더 정확한 태그 찾기 
song_name = element.select('span > a')
song_name
```

```python
# 노래 제목 찾기 - a 태그에서 좀 더 정확한 태그 찾기 
song_name = element.select('div.ellipsis.rank01 > span > a')
song_name
```

```python
# 노래 제목 찾기 - 첫 번째 원소를 선택하기 
song_name = element.select('div.ellipsis.rank01 > span > a')[0]
song_name
```

```python
# 노래 제목 찾기 - 텍스트 부분 추출하기
song_name = element.select('div.ellipsis.rank01 > span > a')[0].text
song_name
```

```python
# 가수 이름 찾기 - 1) a 태그 찾기
singer = element.select('a')
singer
```

```python
# 가수 이름 찾기 - 2) a 태그 찾기 (원하는 정보만 찾기)
singer = element.select('div.ellipsis.rank02 > a')
singer
```

```python
# 가수 이름 찾기 - 3) 첫번째 원소 추출하기
singer = element.select('div.ellipsis.rank02 > a')[0]
singer
```

```python
# 가수 이름 찾기 - 4) 텍스트 추출하기
singer = element.select('div.ellipsis.rank02 > a')[0].text
singer
```

```python
# 앨범제목 찾기
album_name = element.select('div.ellipsis.rank03 > a')[0].text
album_name
```

```python
# 좋아요 숫자 찾기
like_count = element.select('span.cnt')[0].text
like_count
```

```python
# 좋아요 숫자 찾기 - 우리가 원하는 부분 찾기
like_count = element.select('span.cnt')[0].text.split('\n')
like_count
```

```python
# 좋아요 숫자 찾기 - 우리가 원하는 부분 찾기 (리스트에서 세번째 원소 선택하기)
like_count = element.select('span.cnt')[0].text.split('\n')[2]
like_count
```

```python
element_result = [ranking, song_name, singer, album_name, like_count]
element_result
```

```python
# 정보 뭉치에서 정보찾는 것 반복하기
result = []

for element in lst:
    ranking = element.select('span.rank')[0].text
    song_name = element.select('div.ellipsis.rank01 > span > a')[0].text
    singer = element.select('div.ellipsis.rank02 > a')[0].text
    album_name = element.select('div.ellipsis.rank03 > a')[0].text
    like_count = element.select('span.cnt')[0].text.split('\n')[2]

    element_result = [ranking, song_name, singer, album_name, like_count]

    result.append(element_result)
```

```python
result
```

```python
# 보기 좋게 표 형태로 변경
result_df = pd.DataFrame(result)
result_df
```

```python
columns_name = ['랭킹', '제목', '가수', '앨범명', '좋아요 수']
result_df = pd.DataFrame(result, columns=columns_name)
result_df
```

```python
# 10. 엑셀파일로 저장하기
result_df.to_excel('melon_top100.xlsx')
```
